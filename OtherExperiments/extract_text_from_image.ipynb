```python
import os
import easyocr
import pandas as pd

def extract_text_from_images_in_folder(folder_path):
    reader = easyocr.Reader(['en'])  # 'en' for English, add more languages as needed
    texts = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_path = os.path.join(root, file)
                results = reader.readtext(image_path)
                text = ' '.join([result[1] for result in results])
                texts.append({'folder': root, 'filename': file, 'text': text})
    return texts

def save_to_excel(data, excel_path):
    df = pd.DataFrame(data)
    df.to_excel(excel_path, index=False)

# Path to the main folder
main_folder_path = 'D:/Master Thesis/src/images/'  # Update this path

# Path for the Excel file to be saved
excel_path = 'D:/Master Thesis/src/images/claims.xlsx'  # Update this path

# Extract text and save to Excel
print("Extracting text from images and saving to Excel...")
extracted_data = extract_text_from_images_in_folder(main_folder_path)
save_to_excel(extracted_data, excel_path)

print("Extraction and saving to Excel completed.")
```
