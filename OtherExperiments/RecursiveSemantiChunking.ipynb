```python
from llama_index.core import SimpleDirectoryReader

# load documents
documents = SimpleDirectoryReader(input_files=["/content/corpus.txt"]).load_data()
```

```python
from llama_index.core.node_parser import (
    SentenceSplitter,
    SemanticSplitterNodeParser,
)
from llama_index.embeddings.openai import OpenAIEmbedding

import os

os.environ["OPENAI_API_KEY"] = "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"
```

```python
embed_model = OpenAIEmbedding()
splitter = SemanticSplitterNodeParser(
    buffer_size=1, breakpoint_percentile_threshold=95, embed_model=embed_model
)

# also baseline splitter
base_splitter = SentenceSplitter(chunk_size=512)
```

```python
nodes = splitter.get_nodes_from_documents(documents)
```

```python
evidence_chunks = []
```

```python
for i in range(len(nodes)):
  doc = nodes[i].get_content()
  evidence_chunks.append(doc)
```

```python
from semantic_text_splitter import TextSplitter
from tokenizers import Tokenizer
```

```python
token_chunks = []
```

```python
for i in range(len(evidence_chunks)):
  print(i)
  # Maximum number of tokens in a chunk
  max_tokens = 300
  tokenizer = Tokenizer.from_pretrained("bert-base-uncased")
  splitter = TextSplitter.from_huggingface_tokenizer(tokenizer)

  chunks = splitter.chunks(evidence_chunks[i], max_tokens)
  for x in chunks:
    token_chunks.append(x)
```

```python
import os

directory = "/content/chunks/"

# Check if the directory exists
if not os.path.exists(directory):
    # If the directory does not exist, create it
    os.makedirs(directory)


for i, text in enumerate(token_chunks):
    filename = f"chunk{i}.txt"
    filepath = os.path.join(directory, filename)
    with open(filepath, "w") as file:
        file.write(text)
```
